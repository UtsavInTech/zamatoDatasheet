```python
#Reading country code and merging with Zamato Datasheet

countrycode=pd.read_excel('Country-Code.xlsx',engine='openpyxl')
countrycode
```

```python
resturants = resturants.merge(countrycode, on='Country Code')
resturants.head()
```

```python
#Creating region specific Dataframes for Analysis

indiaRes=resturants[resturants['Country']=='India']
singaporeRes=resturants[resturants['Country']=='Singapore']
AusRes=resturants[resturants['Country']=='Australia']
BrazRes=resturants[resturants['Country']=='Brazil']
CanRes=resturants[resturants['Country']=='Canada']
UAERes=resturants[resturants['Country']=='UAE']
UKRes=resturants[resturants['Country']=='United Kingdom']
USRes=resturants[resturants['Country']=='United States']

# 5 resturants of india
indiaRes.head()
```

```python
DelhiTopRes=indiaRes[(indiaRes['City']=='New Delhi') & (indiaRes['Aggregate rating']>4.5)][['Aggregate rating','Restaurant Name','Votes']].sort_values(ascending=False, by ='Votes')
```

```python
DelhiTopRes["Aggregate rating"] = DelhiTopRes["Aggregate rating"].round(1)
```

```python
DelhiTopRes['Aggregate rating']
```

```python
DelhiTopRes.style.apply(lambda x:['background: green' if x.name=='Votes' else 'background: lightsteelblue' for i in x])
```

```python
singaporeTopRes=singaporeRes[(singaporeRes['City']=='Singapore') & (singaporeRes['Aggregate rating']>3.5)][['Aggregate rating','Restaurant Name','Votes']].sort_values(ascending=False, by ='Votes')
```

```python
singaporeTopRes.style.apply(lambda x:['background: pink' if x.name=='Votes' else 'background:lightsteelblue' for i in x])
```

```python
#Objective: Cheap but Best Restaurant in New Delhi
cheap_best_delhi = indiaRes[(indiaRes['City'] == 'New Delhi') & (indiaRes['Aggregate rating'] > 4.0)].sort_values(by=['Aggregate rating', 'Average Cost for two', 'Votes'], ascending=[False, True, False]).head(1)
print("Cheap but Best Restaurant in New Delhi:")
print(cheap_best_delhi[['Restaurant Name', 'Average Cost for two', 'Aggregate rating', 'Votes']])
```

```python
#"Aggregate rating of all the restaurants in each city in a country."
table = pd.pivot_table(data=resturants, index=('Country','City'), values="Aggregate rating")
cm = sns.light_palette("green", as_cmap=True)
table.style.background_gradient(cmap=cm, axis=0)
```

```python
# Objective: Top Restaurant Share by City (Pie Chart)
plt.figure(figsize=(10,10))
indiaRes['City'].value_counts().head().plot(kind='pie',autopct='%1.1f%%')
plt.title("Top Indian Cities by Restaurant Share")
plt.ylabel("")
plt.show()
```

```python
# 13. Objective: Does Rating Influence the Cost? (Boxplot)
plt.figure(figsize=(12,6))
sns.boxplot(x='Aggregate rating', y='Average Cost for two', data=indiaRes)
plt.title("Does Rating Influence the Cost?")
plt.show()
```

```python
# 15. Objective: Top Cuisines in India (Pie Chart)
plt.figure(figsize=(8,8))
indiaRes['Cuisines'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Cuisines in Indian Restaurants")
plt.ylabel("")
plt.show()
```
